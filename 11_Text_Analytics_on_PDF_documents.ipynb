```python
import textract
```

```python
!wget https://nslegislature.ca/sites/default/files/legc/statutes/freedom%20of%20information%20and%20protection%20of%20privacy.pdf
```

```python
# Step 2 - Read the content from the PDF file
text = textract.process("freedom of information and protection of privacy.pdf")
text = str(text)
print("Type of text", type(text))
#print("Show content from pdf \n", text)
```

```python
#Remove some special characters out
import re
text = re.sub("xe2","",text)
text = re.sub("x80","",text)
text = re.sub("x99","",text)
text = re.sub("x0c","",text)
text = re.sub("x93","",text)
text = re.sub("nCase Study","",text)
```

```python
#Explore the pdf content using the wordcloud
from wordcloud import WordCloud
wc = WordCloud().generate(text)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,20))
plt.imshow(wc)
plt.axis("off")
  #plt.title(url)
  #print(url)
plt.show()
```
